## Garbage Detection — Multi-Item YOLOv8 + Gradio

### ML Course Project — Object Detection on Cluttered Trash Scenes

**Task:** Object Detection (multiple items per image, with bounding boxes)
**Model:** YOLOv8 (transfer learning from COCO-pretrained weights)
**Dataset:** TACO — Trash Annotations in Context
**GUI:** Gradio, with boxes drawn directly on the uploaded image

## Runtime tip: Go to Runtime → Change runtime type → T4 GPU before running.


```python
# Install required packages
!pip install ultralytics gradio opencv-python pillow numpy

import torch
import cv2
import numpy as np
import gradio as gr
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Load YOLOv8 model
model = YOLO("yolov8n.pt")

# Train the model
# results = model.train(data="taco_consolidated/data.yaml", epochs=50, imgsz=640, batch=16)
```

```python
# Create Gradio interface
def detect_trash(image):
    results = model(image)
    annotated_image = results[0].plot()
    return annotated_image

# Launch Gradio app
iface = gr.Interface(fn=detect_trash, inputs="image", outputs="image", title="Garbage Detection")
iface.launch(share=True)
```
